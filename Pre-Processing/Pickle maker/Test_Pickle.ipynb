```python
import pickle
import gzip
import os
```

```python
# Uncomment the file you need to check
# pickle_dest = "/Dataset/Pickles/excel_dataTest.pickle"
pickle_dest = "/Dataset/Pickles/excel_dataVal.pickle"
# pickle_dest = "/Dataset/Pickles/excel_dataTrain.pickle"
```

```python
# pickle_dest = os.getcwd() + pickle_dest

def view_gzip_pickle_file(file_path):
    try:
        with gzip.open(os.getcwd() + file_path, 'rb') as file:
            data = pickle.load(file)
            print("Contents of the gzip-compressed pickle file:")
            print(data)
            return data
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"Error: {e}")
```

```python
# Example usage:
out = view_gzip_pickle_file(pickle_dest)

```
